```python
# Guessing word/password game
import random as r
easy_words=["apple","banana","game","glass","book"]
Medium_words=["python","bottom","planet","upper","laptop"]
hard_words=["computer","umbrella","electronic","elephant","mountain"]
print("enter difficulty level (easy,medium,hard)")
level=input("enter your level: ").lower().strip()
if level=="easy":
  word=r.choice(easy_words)  # computer ne random word choices kiya hai
elif level=="medium":
  word=r.choice(Medium_words)
elif level=="hard":
  word=r.choice(hard_words)
else:
  print("Invalid input by default easy level:")
  word=r.choice(easy_words)

attempt =0
print("\n Guess the secret password:")

while True:
  guess =input("enter you guess word: ").lower().strip()
  attempt +=1
  if guess==word:
    print(f" Congratulation! you guess it in {attempt} attempt")
    break

  hint=""
  for i in range(len(word)):
    if i<len(guess) and guess[i]==word[i]:
      hint +=guess[i]
    else:
      hint += "*"
  print(f"Hint: {hint}")

print("Game Over")


```
