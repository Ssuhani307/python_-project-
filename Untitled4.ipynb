```python
#______Define the manu of restaurants______
print("Welcome to SUHANI'S restaurant. Here's the menu :")
menu={"Pizz": 40 ,"Pasta":50,"Burger":60,"Salad":45,"Coffee":76 }
order_total=0
while True:
  #print menu
  for key,value in menu.items():
    print(f"{key}-₹{value}")
  #input to guest for ordering
  t=input("you ordered something (or type exist to finish): ").capitalize().strip()
  if t=="Exist":
    break
  elif t in menu:
    order_total+=menu[t]
    print(f"order of {t} has been added")
  else:
   print(f"Sorry your item {t} is not available yet!")
#Total amount you pay
print(f"The total amount of item to pay is ₹{order_total}")
print("Thank you for ordering. Visit again")
```
