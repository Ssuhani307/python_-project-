```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 70)
y = np.sin(np.exp(x))

plt.plot(x , y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(e^x)')

```

```python
import matplotlib.pyplot as plt
college_type= ['SVNIT Surat ','SVNIT Surat ','SVNIT Surat' ,'NIT Warangal','NIT Warangal','NIT Warangal']

departments = ['CSE', 'ECE', 'Mechanical', 'Civil', 'CSE', 'ECE', 'Mechanical', 'Civil']
total_students = [120, 100, 90, 80, 150, 130, 110, 90]
placed_students = [110, 85, 70, 60, 140, 120, 90, 75]

# (a)
svnit_matrix = np.array([[90, 70], [80, 60]])
nitw_matrix = np.array([[110, 90], [90, 75]])
print(svnit_matrix)
print(nitw_matrix)
 #(b)
result_matrix = np.dot(svnit_matrix, nitw_matrix.T)
print("Matrix Multiplication Result:\n", result_matrix)

# (c)
low_placement_departments = []

for i in range(len(departments)):
    if placed_students[i] <= 70:
        low_placement_departments.append((college_type[i], departments[i]))

print("Departments with 70 or fewer placements:")
for college_type, departments in low_placement_departments:
    print(college_type,"-",departments)

```

```python
import numpy as np

#given matrixes A and B
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# (a) Find the sum of the two matrices A + B
sum_matrix = A + B
print("Sum of A and B:")
print(sum_matrix)

# (b) Find the difference of the two matrices A - B
difference_matrix = A - B
print("\nDifference of A and B:")
print(difference_matrix)
#(c) find the  min and max values from matrix A

min_value = A.min()
max_value = A.max()

print("\nMinimum value in matrix A:", min_value)
print("Maximum value in matrixA:",max_value)


```
