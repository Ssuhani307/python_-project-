```python
# create  random library
import random
#create list
subject =["Ek confused student","Ek angry teacher","Ek overconfident uncle","Ek jugaadu friend","Ek sleepy teacher","Ek emotional AI robot","Ek angry mom","Ek silent topper","Ek lazy cat","Ek desi engineer","Ek overthinking girl"]
action=["Chai ko motivation bata raha hai","Attendance ke liye selfie le raha hai","Wifi password ke liye dua kar raha hai","Crush ko impress karte hue fail ho gaya",
"Project ko bhagwan se mila raha hai","Sleep mode me lecture attend kar raha hai","Homework ko “creative writing” bol raha hai","Maggi ko 5-star dish bana diya","Internships ke liye astrologer se mil raha hai","Library me comedy show chala diya"]
place=["Canteen me","Metro me","College viva ke time","Gym me","Library me","Marriage function me","Exam hall me","Hostel ke terrace par","Zoom meeting me","Classroom ke last bench par"]
#start headline genration to loop
while True:
  s=random.choice(subject)
  a=random.choice(action)
  p=random.choice(place)
  print("BREAKING HEADLINES:")
  print(f"{s} {a} {p}")
  r=input("do yo want to generate other fuuny headlines:(Type yes or no)").capitalize().strip()
  if r=="No":
    break


```

```python
# calculator
HISTORY_File ="history.txt"
def show_history_file():
  print("History of calculations:")
  with open(HISTORY_File,"r") as f:
    n=len(f.readline())    #read each line and calculate len of files line
    if n==0:
      print("empty history")
    else:
      for line in f:
        print(line)
    f.close()


def clear_history ():
  with open(HISTORY_File,"w") as f:
    #f.write("")
    print("History cleared")
  f.close()

def save_history (equation,result):
  with open(HISTORY_File,"a") as f:
    f.write(f"{equation} = {result}\n")
  f.close()

def calculate(user):
  part=user.split()
  if len(part)!=3:
    print("invalid input")
    return None
  num1=float(part[0])
  num2=float(part[2])
  if part[1]=="+":
    result=num1+num2
  elif part[1]=="-":
    result=num1-num2
  elif part[1]=="*":
    result=num1*num2
  elif part[1]=="/":
    if num2==0:
      print("Error: Division by zero")
      return None
    result=num1/num2
  elif part[1]=="**":
    result=num1**num2
  else:
    print("invalid operator (enter +,-,*,/,**)")
    return None
  if int(result)==result:
    result=int(result)
  print(result)
  save_history(user,result)


def main():
    print("______Simple Calculator______ (type history,clear,exit)")
    while True:
      user=input("Enter your equation:or type (i.e 4 + 5(operation is +,-,*,/,**)): ")
      if user.lower()=="exit":
        print("good bay!")
        break
      elif user.lower()=="history":
        show_history_file()
      elif user.lower()=="clear":
        clear_history()
      else:
        calculate(user)
main()

```
